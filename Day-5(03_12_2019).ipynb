# Operator


1. Arithmetic Operators-----> +,-,*,/,%,**, // 
2. Logical Operators -------> and , or , not
3. Relational Operators-----> ==,!=,>,<,>=,<=
4. Membership Operators-----> in , not in
5. Conditional Statements---> if , else, elif


```
a = 10
b = 5
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
print(a//b)
```

```
# Logical Operators
i = 10
j = 20
print(i>1 and i<20)
print(i>12 or i<33)
print(not i)
```

```
# relational operators
print(i==j)
print(i!=j)
print(i>y)
print(i<j)
print(i>=j)
print(i<=j)
```

```
# Membership Operators
str1= 'gitam'
print('a' in str1)
print('a' not in str1)
```
