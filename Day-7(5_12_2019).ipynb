```
# Function for finding Factorial of a number
def factorial(n) :
  factorial = 1
  f = 1
  for i in range(1,n+1) :
    if f<=n :
      factorial = factorial*i
      f = f-1
  return factorial
print(factorial(5))
print(factorial(10))
```
