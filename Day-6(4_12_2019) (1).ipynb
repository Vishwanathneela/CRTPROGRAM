# Accept three subject marks of a student 
#####1.if the student given 30 marks in all the subjects then calculate average else 
print fail.
#####2.if average is greater than 75 print A grade.
#####3.if average lies B/w 65-75 print B grade.
#####4.if average lies B/w 45-65 print C grade.
#####5.if average lies B/w 30-45 print D grade.
#####6.if average is less than 30 print F grade.

```
math = int(input("Enter the marks in Mathematics"))
phy = int(input("Enter the marks in Physics"))
chem = int(input("Enter the marks in Chemistry"))
avg = (math+phy+chem)/3
if math>=30 and phy>=30 and chem>=30 :
  if avg>75 :
    print("The average marks in three subjects is",avg)
    print("The student has aquired A grade")
  elif avg>65 and avg<=75 :
    print("The average marks in three subjects is",avg)
    print("The student has aquired B grade")
  elif avg>45 and avg<=65 :
    print("The average marks in three subjects is",avg)
    print("The student has aquired C grade")
  elif avg>30 and avg<=45 :
    print("The student has aquired D grade")
else :
  print("Fail and student has aquired F grade")
```

# Loop

2. For Loop



```
# While Loop
# Accept a number(N) and print all the values from 0 to N
n = int(input("Enter the value of N"))
i = 1
while i<=n :
  print(i,end='  ')
  i=i+1
```

```
# Printing Even Natural numbers
n = int(input("Enter the value of N"))
i = 1
while i<=n:
  if i%2==0:
    print(i,end=' ')
  i=i+1
```

```
# Printing Sum of Even numbers in the given range
n = int(input("Enter the Number"))
i = 1
sum = 0
while i<=n:
  if i%2==0:
    sum = sum+i 
  i=i+1
print(sum)
```

```
# Extracting the Digits from the input
n = int(input("Enter the number"))
while n!=0 :
  print(n%10,end='  ')
  n=n//10
```

```
# Palindrome 
n = int(input("Enter the Number"))
rev = 0
temp = n
while n!=0 :
  r=n%10
  rev=rev*10+r
  n=n//10
if rev == temp :
  print("palindrome")
else :
  print("Not Palindrome")
```

```
# For Loop
n = int(input("Enter the value"))
for i in range(1,n+1,3) :
  print(i)
```

```
name = input("enter your name")
for i in name:
  print(i)
```
